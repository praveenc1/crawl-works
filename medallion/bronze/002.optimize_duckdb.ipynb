# Assume that the environment is already setup

```python
import duckdb
```

```python
DB_NAME_FULL='data/cc_index.duckdb'
DB_NAME_15_PCT= 'data/cc_index_15pct_sample.duckdb'
DB_NAME_P1_PCT= 'data/cc_index_p1pct_sample.duckdb'

DB_NAME= DB_NAME_P1_PCT
```

```python
con = duckdb.connect(DB_NAME)
```

```python
def sql(query):
  return con.execute(query).df()  
```

```python
sql('show tables')
```

```python
sql('select count(*) from cc_index_sample')
```

```python
sql('select distinct content_languages from cc_index_sample limit 20')
```

```python
sql("select distinct content_languages from cc_index_sample where content_languages like '%tel%' limit 20")
```

```python
[print(url) for url in sql("select * from cc_index_sample where content_languages like '%tel%' limit 200")['url']]
```

```python
sql("""
    select 
      url_host_name as host, 
      count(url_host_name) as count 
    from 
      cc_index_sample 
    where 
      content_languages like '%tel%'
    group by 
      host
    order by
      count
    desc
""")
```
